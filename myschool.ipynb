# Summary

* This report based on the past 11 years (from 2008 to 2018) data from `myschool.com.au`. 
* It contains 2000+ school in VIC, which include all primary/secondary/combined/special schools.
* The following graph shows the data of survey, you can use this report to understand the distribution of schools according to ICSEA(Index of Community Socio- Educational Advantage) score.
* Quick Takeway
  * We can validate the ICSEA is obey the normal distribution, which is centered at 1000
  * There are 4 types of school: Primary, Secondary, Combined, Special.
  * The Government Primay school is good choice for Primary.
  * The Non-Government Combined school is good choice for both Primary and Secondary.
  * The Government Combined school is not a good choice
  * The Non-Government Secondary school is also not a good choice.

## Government Primary Rank top 25 for 2018

# The following analyse is based on the past 11 years myschool data

## Government Primary Rank top 20 for the average of last 11 years

```python
pd.read_sql_query('''
SELECT name, suburb, year_range, avg(icsea) as icsea_avg, stddev_samp(icsea) 
FROM myschool2 
where sector='Government' and school_type='Primary' and icsea>1000
group by name, suburb, year_range
having count(icsea)=11
order by icsea_avg desc 
limit 20;''', cnx)
```

## Government Primary Rank top 20 for the average of last 5 years

```python
pd.read_sql_query('''
SELECT name, suburb, year_range, avg(icsea) as icsea_avg, stddev_samp(icsea) 
FROM myschool2 
where sector='Government' and school_type='Primary' and icsea>1000 and year>2013
group by name, suburb, year_range
having count(icsea)=5
order by icsea_avg desc 
limit 20;''', cnx)
```
